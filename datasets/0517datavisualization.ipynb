```python
  
import h5py
from IPython import display
import matplotlib.pyplot as plt
import numpy as np

with h5py.File('test_sample.h5', 'r') as h5f:
    print(h5f["test1"].keys())
    left_eye = h5f["test1/left-eye"][:]
    right_eye = h5f["test1/right-eye"][:]
    eye_region = h5f["test1/eye-region"][:]
    face = h5f["test1/face"][:]
    face_landmarks = h5f["test1/face-landmarks"][:]
    head = h5f["test1/head"][:]
```

## Single group of case

```python
idx = 0
l1 = left_eye[idx]
r1 = right_eye[idx]
f1 = face[idx]
fl1 = face_landmarks[idx]
h1 = head[idx]
er1 = eye_region[idx]
```

## Batch visualization

```python
def plotEyeFace(start,ncol):
    fig = plt.figure(figsize=(ncol*2, 6))
    for i in range(start,start+ncol):
        loc = i+1-start
        plt.subplot(3, ncol, loc)
        plt.imshow(left_eye[i], cmap='gray')
        loc+=ncol
        plt.subplot(3, ncol, loc)
        plt.imshow(right_eye[i], cmap='gray')
        loc+=ncol
        plt.subplot(3, ncol, loc)
        plt.imshow(face[i], cmap='gray')
        plt.plot(face_landmarks[i,:,0],face_landmarks[i,:,1],'ro')

    plt.show()

def plotEyeRegion(start,ncol):
    fig = plt.figure(figsize=(ncol*2, 2))

    for i in range(start,start+ncol):
        plt.subplot(2, ncol/2, i+1-start)
        plt.imshow(eye_region[i], cmap='gray')
    plt.show()
```

```python
start=5
ncol=10
plotEyeFace(start,ncol)
plotEyeRegion(start,ncol)
```
